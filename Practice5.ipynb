```python
import json
import os
import subprocess
import re
import anthropic


client = anthropic.Anthropic(api_key="")

with open("challenge.json", "r") as f:
  challenge = json.load(f)


with open("my_first_pwnie.py", "r") as f:
  source = f.read()


system_prompt =f"""

You are an expert in solve the CTF challanges

challenge: {challenge["name"]}

description: {challenge["description"]}
Files available : {challenge["files"]}

source code{source}



Your task is analyse the challenges and let me know of the vulnearbality and how to soleve it and let mw know how to find the flag

be perciese and give the source code

and after provude a python code that I can manually solve it
"""


response = client.messages.create(
    model="claude-sonnet-4-20250514",
    max_tokens = 2048,
    messages =[{"role":"user", "content": system_prompt}]

  )


print(response.content[0].text)
```
